```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sales.csv')
```

```python
df.head()
```

```python
X = df.drop(['Price'],axis=1)
```

```python
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
```

```python
X_train = sc.fit_transform(X_train)
```

```python
X_test = sc.transform(X_test)
```

```python
from tensorflow import keras
from tensorflow.keras.layers import Dense
```

```python
model = keras.Sequential([
    
        Dense(5,activation='relu',input_dim=5),
        Dense(5,activation='relu'),
        Dense(1,activation=keras.activations.linear)
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error',metrics=['mse'])
```

## [Get started with TensorBoard](https://www.tensorflow.org/tensorboard/get_started)

```python
# Load the TensorBoard notebook extension
%load_ext tensorboard
log_folder = 'logs'
```

```python
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
```

```python
callbacks = [
            EarlyStopping(patience = 30,mode='min',monitor='val_loss'),
            TensorBoard(log_dir=log_folder)
    
            ]
```

```python
history = model.fit(X_train,y_train,epochs=800,validation_split = 0.2,callbacks=callbacks)
```
