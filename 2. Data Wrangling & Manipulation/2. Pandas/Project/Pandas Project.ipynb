```python
# Create a dataframe called sectors that contains the count of the
# number of times a sector appears in the dataset
sectors = df.groupby('sector').id.count().reset_index()
sectors.head()
```

```python
# Rename the id column to count
sectors.rename(columns = {'id':'count'}, inplace=True)
sectors
```

```python
# Create a dataframe called sector_loan that contains the count of the loan
# amounts per sector
sector_loan = df.groupby('sector').loan_amount.count().sort_values(ascending=False).reset_index()
sector_loan
```

```python
# Create a dataframe called usa that contains data from USA only
usa = df[df.country == 'United States']
usa.head(1)
```

```python
# Create a pivot table that shows the sum of loan amounts per sector
usa.pivot_table(columns='sector', values=['loan_amount', 'lender_count'], aggfunc='sum')
```

```python
# Load in the themes data as themes
themes = pd.read_csv('loan_theme_ids.csv')
themes.head()
```

```python
# Display the columns in the themes dataset
themes.columns
```

```python
# Display the columns in df
```

```python
# Merge the themes data frame with df
# and display the merged data frame
m = pd.merge(df, themes, on='id')
m.head()
```

```python
# Print the unique loan theme types
m['Loan Theme Type'].unique()
```

```python
# Which are the most prominent loan themes by id count
r = m.groupby('Loan Theme Type').id.count().reset_index().sort_values('id', ascending=False).head(5)
r.rename(columns={'id':'Count'}, inplace=True)
r
```
